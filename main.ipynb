```python
import pandas as pd

url = "https://www.baseball-reference.com/teams/SEA/2025-schedule-scores.shtml"

tables = pd.read_html(url)

len(tables)
```

```python
game_logs = tables[0]

game_logs.head()
```

```python
game_logs = tables[0]

game_logs.head()
```

```python
# Remove repeated header rows if they exist
game_logs = game_logs[game_logs["Gm#"] != "Gm#"].copy()

# Convert runs to numeric
game_logs["R"] = pd.to_numeric(game_logs["R"], errors="coerce")
game_logs["RA"] = pd.to_numeric(game_logs["RA"], errors="coerce")

# Create home/away column
game_logs["home_away"] = game_logs["Unnamed: 4"].apply(
    lambda x: "Away" if x == "@" else "Home"
)

# Keep useful columns
mariners_games = game_logs[["Date", "Opp", "home_away", "R", "RA", "W/L"]].copy()

mariners_games.head()
```

```python
mariners_games.to_csv("mariners_2025_game_logs.csv", index=False)
```

```python
import matplotlib.pyplot as plt

mariners_games["R"].plot(kind="hist", bins=10)

plt.title("Distribution of Mariners Runs Scored Per Game")
plt.xlabel("Runs Scored")
plt.ylabel("Number of Games")
plt.show()
```

```python
mariners_games.groupby("home_away")["R"].mean()
```

```python
mariners_games.groupby("home_away")["R"].mean().plot(kind="bar")

plt.title("Average Runs Scored: Home vs Away")
plt.xlabel("Location")
plt.ylabel("Average Runs Scored")
plt.show()
```

```python
# Basic summary statistics
mariners_games["R"].describe()
```

```python
# Average runs by opponent
runs_by_opponent = mariners_games.groupby("Opp")["R"].mean().sort_values(ascending=False)

runs_by_opponent.head(10)
```

```python
# Chart: top 10 opponents by average runs scored
runs_by_opponent.head(10).plot(kind="bar")

plt.title("Top 10 Opponents by Average Mariners Runs Scored")
plt.xlabel("Opponent")
plt.ylabel("Average Runs Scored")
plt.show()
```

```python
# Average runs in wins vs losses
mariners_games.groupby("W/L")["R"].mean()
```

```python
mariners_games.groupby("W/L")["R"].mean().plot(kind="bar")

plt.title("Average Runs Scored in Wins vs Losses")
plt.xlabel("Game Result")
plt.ylabel("Average Runs Scored")
plt.show()
```

## Initial EDA Findings

The first dataset from Baseball Savant is player-level batting data. It shows individual Mariners offensive statistics such as batting average, slugging percentage, on-base percentage, home runs, strikeouts, walks, and plate appearances.

Because this dataset is player-level, it does not include game date, opponent, home/away status, or runs scored by game. To answer our project question about run production, we added a separate game-level Mariners dataset.

The game-level dataset allows us to analyze:
- runs scored per game
- home vs. away scoring differences
- scoring by opponent
- scoring in wins compared to losses

This gives our project a stronger structure because the response variable, runs scored, is now measured at the game level.

## Data Merge Note

The Savant dataset and the game log dataset could not be directly merged because they are recorded at different levels of analysis. The Savant dataset is player-level season data, with one row per player, while the game log dataset is game-level data, with one row per Mariners game.

Because there is no shared merge key such as game date, game ID, or opponent, a direct pandas merge would not be meaningful. Instead, the Savant data will be used to summarize player-level offensive performance, while the game log data will be used for game-level run production analysis.
